### Mean, Median, Mode

#### Mean

The mean of a dataset, also known as the average, is the most common measure of central tendency that represents the arithmetic average of a set of values. It is calculated by summing up all the values in the dataset and then dividing the sum by the total number of values.

In mathematical terms, if we have a dataset $X$ with $n$ values, the mean $(μ)$ is calculated as follows:
$$ \mu = \frac{1}{n} \sum_{i=1}^{n} X_i $$

Here:
- $n$ is the total number of values in the dataset.
- $X_{i}$ represents each individual value in the dataset
- $\sum_{i=1}^{n}$ denotes the summation of all values from $i=1$ to $n$

For a series we can get the mean by calling .mean() method

```python
print("radius mean column mean:", df['radius_mean'].mean())
```

Similarly, we can get the mean of the whole dataframe

#### Median

The median is often a better measure of central tendency when we have extreme outliers. The median is the value in the middle after we sort the data. This is why outliers do not influence the median as much as they do the mean. If the number of observations $n$ is even, we have to take the mean of the two middle values. We calculate for a zero based index.

$$
M = \begin{cases} 
X_{\frac{n+1}{2}} & \text{if } n \text{ is odd} \\[10pt]
\frac{X_{\frac{n}{2}} + X_{\frac{n}{2} + 1}}{2} & \text{if } n \text{ is even}
\end{cases}
$$


```python
df[numerical_columns].median()
```

#### Mode

The mode of a dataset refers to one or more values that occur with the highest frequency, making it a measure of central tendency particularly relevant in situations with numerous repeated data points. This statistical measure identifies the most common or frequently occurring values in a series. 

$$
Mo = \text{{argmax}}(f(X_i))
$$

where $f(X_i)$ represents the frequency of the $i-th$ value in the dataset. The mode is the value or values for which this frequency is maximized.

```python
df.mode().iloc[0]
```

### Standard Deviation

Standard deviation is a measure of the amount of variation or dispersion in a set of values. It quantifies how much individual values in a dataset differ from the mean (average) of the dataset. In other words, it provides a measure of the spread of values around the mean.

The standard deviation $(σ)$ is calculated using the following formula:
$$ \sigma = \sqrt{\frac{\sum_{i=1}^{n}(x_i - \mu)^2}{n}} $$

σ represents the standard deviation.
- $n$ is the number of data points in the dataset.
- $x_i$ represents each individual data point.
- $μ$ is the mean of the dataset.


```python
df[numerical_columns].std()
```

### Quantiles

1. **Q1 (First Quartile):** It represents the value below which 25% of the data falls. In other words, Q1 is the 25th percentile of the dataset.
2. **Q2 (Second Quartile):** It represents the median of the dataset. This is the value that separates the higher half from the lower half of the data.
3. **Q3 (Third Quartile):** It represents the value below which 75% of the data falls. Q3 is the 75th percentile of the dataset.

```python
data = df['radius_mean']
data.describe()
```

```python
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print('*********************************')

# Another method
sorted_data = np.sort(data)

Q2 = np.median(sorted_data)

# Calculate Q1
lower_half = sorted_data[sorted_data <= Q2]
Q1 = np.median(lower_half)

# Calculate Q3
upper_half = sorted_data[sorted_data >= Q2]
Q3 = np.median(upper_half)

print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
```

## Distributions

In statistics and probability theory, a distribution refers to the set of all possible values and their corresponding probabilities (or frequencies) for a particular random variable or set of variables. It describes how the values of a variable are spread or distributed across the range of possible outcomes.

### Gaussian Distribution

The Gaussian distribution, also known as the normal distribution, is a continuous probability distribution that is symmetric bell-shaped curve centered around the dataset's mean $(μ)$ and width of the curve is defined by the variance (standard deviation $(σ)$). The probability density function features two elongated tails on each end. The more rare the instance — the further it is from the center — the more likely it is to be an outlier or an anomaly.

#### Probability Density Function (PDF) Equation:

The probability density function (PDF) of the Gaussian distribution is given by:

$$ f(x; \mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x - \mu}{\sigma}\right)^2} $$

- $x$ is the variable.
- $μ$ is the mean.
- $σ$ is the standard deviation.
- $e$ is the base of the natural logarithm.


#### Usage in Machine Learning:

1. Feature Scaling:
    Techniques like Z-score normalization involve transforming data to have a mean of 0 and a standard deviation of 1, assuming a Gaussian distribution. This is commonly used in feature scaling.

2. Generative Models:
   Gaussian distributions are fundamental in generative models like Gaussian Mixture Models (GMMs) and Naive Bayes classifiers. GMMs assume that data is generated from a mixture of several Gaussian distributions.

3. Anomaly Detection:
   Gaussian distribution is often used in anomaly detection where normal instances are expected to follow a Gaussian distribution, and anomalies are deviations from this pattern.
   
4. Probability Density Estimation:
   Gaussian distribution is used in estimating the probability density function of a dataset. Kernel Density Estimation (KDE) is an example where Gaussian kernels are commonly used.

```python
import matplotlib.pyplot as plt

mu = 0     # mean
sigma = 1  # standard deviation

# Generate data points using the Gaussian PDF equation
x = np.linspace(-5, 5, 1000)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Plot the Gaussian distribution
plt.figure(figsize=(8, 6))
plt.tight_layout()
plt.plot(x, y, label='Gaussian Distribution', color='blue')
plt.title('Gaussian Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import norm

y = norm.pdf(x, mu, sigma)  # Probability Density Function (PDF) for Gaussian distribution

plt.figure(figsize=(8, 6))
plt.tight_layout()
plt.plot(x, y, label='Gaussian Distribution', color='blue')
plt.title('Gaussian Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

# Probability

## Probability of Independent Events
First an exploration of independent events. This is where one event doesn't effect the other. If we flip a fair coin, the probability that the comes up heads or tails is for both 50%. Simply by intuition we know:

$$P(heads) = 0.5,\ P(tails)=0.5$$

If we flip a biased (loaded, unfair) coin, the odds change. Let's say we have a coin that comes up tails 75% of the time.

$$P(tails) = 0.75,\ P(heads)=1.0-0.75=0.25$$

This leads to a basic probability law that the probability of $A$ happening is equal to the probability of $1$ minus $A$ *not* happening, or $A^{'}$ which is called the complement of $A$.

$$P(A) = 1.0 -P(A^{'})$$



## Bayes' Theorem

Bayes' theorem is a fundamental formula in probability theory that describes the probability of an event based on prior knowledge of conditions that might be related to the event.

It is said that the Reverend Thomas Bayes developed his rule on inverse probability while he was trying to prove the existence of God somewhere around 1740. He came up with a method for calculating the probability of an event occurring given that another event has occurred. Starting out with the prior probability (or believe) $P(A)$, when given a likelihood $P(B | A)$ and evidence $P(B)$ we arrive at the posterior probability $P(A | B)$.

Bayes Rule proves to be a powerful tool and is widely used in diverging areas like economics, artificial intelligence, medicine, journalism, military, just to name a few. Most spam filters use Bayes Rule in one way or another. The Bayes' Theorem formula is, posterior = likelihood times prior, over evidence:

$$ P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B)} $$

Where:
- $P(A∣B)$ is the probability of event A given that event B has occurred.
- $P(B∣A)$ is the probability of event B given that event A has occurred.
- $P(A)$ is the prior probability of event A.
- $P(B)$ is the prior probability of event B.

### Example: Medical Test for a Rare Disease

Where Bayes' Rule shines is when we have evidence (for instance a test result) which is not 100% accurate. Suppose you have a rare disease, and there is a diagnostic test for this disease. The prevalence of the disease in the general population is 0.1%  (proportion of people having cancer of the total population). The diagnostic test has probability of returning a positive result given the person has cancer is of 95% (called the [sensitivity](https://en.wikipedia.org/wiki/Sensitivity_and_specificity)) and a [specificity](https://en.wikipedia.org/wiki/Sensitivity_and_specificity) of 90% (The specificity denotes the probability of having a negative test result given the person has no disease).



Definitions:
- $A$: Having the disease
- $B$: Positive test result

Given Probabilities:
- $P(A)$: Prevalence of the disease = 0.001 (0.1%)
- $P(B∣A)$: Sensitivity of the test = 0.95 (95%)
- $P(¬B∣¬A)$: Specificity of the test = 0.90 (90%)

<img src="bayes.png" height=400px>

Here, 
$P(B)$ is the total probability of observing a positive test result, and it's composed of two parts:

**True Positive Probability ($P(B∣A)⋅P(A)$)**:

- $P(B∣A)$: The probability of testing positive given that the person has the disease (sensitivity of the test).
- $P(A)$: The prevalence of the disease in the population.
  
Together, $P(B∣A)⋅P(A)$ represents the probability of both having the disease and testing positive. This is the contribution to $P(B)$ from individuals who truly have the disease.

**False Positive Probability ($P(B∣¬A)⋅P(¬A)$)**:

- $P(B∣¬A)$: The probability of testing positive given that the person does not have the disease (false positive rate).

- $P(¬A)$: The complement of the prevalence of the disease, representing the probability of not having the disease.
- 
Together, $P(B∣¬A)⋅P(¬A)$ represents the probability of not having the disease but still testing positive. This is the contribution to $P(B)$ from individuals who do not have the disease.

Calculation:
$$

\begin{equation*}
    P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A)}
\end{equation*}
​$$

$$
\begin{align*}
    P(A|B) & = \frac{0.95 \cdot 0.001}{(0.95 \cdot 0.001) + (0.10 \cdot 0.999)} & \approx 0.0094 \quad (0.94\%)
\end{align*}
$$

```python
# Given Probabilities
p_a = 0.001  # Prevalence of the disease
p_b_given_a = 0.95  # Sensitivity of the test
p_b_given_not_a = 0.1  # 1 - Specificity of the test

p_b = ((p_b_given_a * p_a) + (p_b_given_not_a * (1 - p_a)))

print("Probability of having a positive test:", str(p_b*100)+"%")

# Bayes' Theorem Calculation
p_a_given_b = (p_b_given_a * p_a) / p_b

print("Probability of having the disease given a positive test result:", str(p_a_given_b*100)+'%')

p_not_a_given_b = (p_b_given_not_a * (1 - p_a)) / p_b

print("Probability of not having the disease given a positive test result:", str(p_not_a_given_b*100)+'%')
```
